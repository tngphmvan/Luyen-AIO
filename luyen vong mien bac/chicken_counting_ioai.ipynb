## **5. Chấm Điểm**

Bạn sẽ được chấm điểm dựa trên sai số tương đối trung bình của mô hình. Sai số tương đối được định nghĩa bởi:

$$
\text{Sai Số Tương Đối} = \frac{|y_i - \hat{y}_i|}{|y_i|}
$$

trong đó $y_i$ là tổng mật độ thực cho mẫu thứ $i$, và $\hat{y}_i$ là tổng mật độ dự đoán cho mẫu thứ $i$.

Điểm số cuối cùng của bạn trước khi chuẩn hóa sẽ được tính dựa trên sai số tương đối trung bình, như sau:

$$
\text{Điểm Số} = \exp(-\frac{1}{n} \sum_{i=1}^{n} \frac{|y_i - \hat{y}_i|}{y_i})
$$
